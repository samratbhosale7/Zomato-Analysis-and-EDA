```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
# lets save above cell in a variable
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# observations : 
1. when the rating is between 4.5 to 4.9 --- excellent
2. when the rating is between 4.0 to 4.5 --- very good
3. when the rating is between 3.5 to 3.9 --- good
4. when the rating is between 3.0 to 3.4 --- average
5. when the rating is between 2.5 to 2.9 --- average
6. when the rating is between 2.0 to 2.4 --- poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green'])
```

observation:
1. not rated count is very high
2. Maximum no of ratings are betweeen 2.5  and 3.4


```python
#count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green'])
```
